```python
#renombrar columnas con minusculas
df.columns = df.columns.str.lower()
#renombrar columna bhk a number of rooms
df.rename(columns={'bhk': 'number of rooms'}, inplace=True)
```

```python
df['Floor'].unique()
```

```python
def parse_floor(value):
    if pd.isnull(value):
        return np.nan, np.nan
    value = value.strip()
    if 'Ground' in value:
        total = int(value.split('out of')[1].strip()) if 'out of' in value else np.nan
        return 0, total
    elif 'Upper Basement' in value:
        total = int(value.split('out of')[1].strip())
        return -1, total
    elif 'Lower Basement' in value:
        total = int(value.split('out of')[1].strip())
        return -2, total
    elif 'out of' in value:
        parts = value.split('out of')
        floor = int(parts[0].strip())
        total = int(parts[1].strip())
        return floor, total
    elif value.isdigit():
        return int(value), np.nan
    else:
        return np.nan, np.nan

# Crear columnas 'floor_number' y 'total_floors' a partir de la columna 'Floor'
df[['floor_number', 'total_floors']] = df['Floor'].apply(lambda x: pd.Series(parse_floor(x)))
```
